## Задание 1

Даны 2 переменных, в которых хранятся строки произвольной длины: *phrase_1* и *phrase_2*.  
Напишите код, который проверяет какая из этих строк длиннее.

Примеры работы программы:  
1.  

```
phrase_1 = 'Насколько проще было бы писать программы, если бы не заказчики'

```

Результат:

```

phrase_2 = 'Насколько проще было бы писать программы, если бы не заказчики'
```

3.     

```
phrase_1 = 'Насколько проще было бы писать программы, если бы не заказчики'
phrase_2 = 'Насколько проще было бы писать программы, если бы не заказчики'
```

Результат:

````python
# Данные строки
phrase_1 = '640Кб должно хватить для любых задач. Билл Гейтс (по легенде)'
phrase_2 = 'Насколько проще было бы писать программы, если бы не заказчики'
# Получаем длины строк
len_1 = len(phrase_1)
len_2 = len(phrase_2)

# Сравниваем и выводим результат
if len_1 > len_2:
    print("Фраза 1 длиннее фразы 2")
elif len_1 < len_2:
    print("Фраза 2 длиннее фразы 1")
else:
    print("Фразы равной длины")
````

## Задание 5 (необязательное)

Дана переменная, в которой хранится шестизначное число (номер проездного билета). Напишите программу, которая будет определять, является ли данный билет "счастливым". Билет считается счастливым, если сумма первых трех цифр совпадает с суммой последних трех цифр номера.   

Примеры работы программы:  
1. ```number = 123456```

Результат:  
`Несчастливый билет`  
  
2. ```number = 123321```

Результат:  
`Счастливый билет`


````python
# Задаем номер билета
number = 123321  

# Превращаем число в строку, чтобы работать с отдельными цифрами
number_str = str(number)

# Берем первые три цифры и последние три цифры
first_three = number_str[:3]  # цифры с 0 по 2 индекс
last_three = number_str[3:]   # цифры с 3 по 5 индекс

# Считаем суммы
sum_first = int(first_three[0]) + int(first_three[1]) + int(first_three[2])
sum_last = int(last_three[0]) + int(last_three[1]) + int(last_three[2])

# Сравниваем суммы
if sum_first == sum_last:
    print("Счастливый билет")
else:
    print("Несчастливый билет")
````

````python
# Доработано: альтернативное решение через математические операции
number = 123331  

# Извлекаем цифры с помощью целочисленного деления и остатка от деления
num_1 = number // 100000          # первая цифра (сотни тысяч)
num_2 = number // 10000 % 10      # вторая цифра (десятки тысяч)
num_3 = number // 1000 % 10       # третья цифра (тысячи)
num_4 = number // 100 % 10        # четвертая цифра (сотни)
num_5 = number // 10 % 10         # пятая цифра (десятки)
num_6 = number % 10               # шестая цифра (единицы)

# Считаем суммы
sum_first = num_1 + num_2 + num_3
sum_last = num_4 + num_5 + num_6

# Выводим результат
print(f"Номер билета: {number}")
print(f"Первые три цифры: {num_1}{num_2}{num_3}, сумма = {sum_first}")
print(f"Последние три цифры: {num_4}{num_5}{num_6}, сумма = {sum_last}")

if sum_first == sum_last:
    print("Результат: Счастливый билет!")
else:
    print("Результат: Несчастливый билет")
````

## Задание 6 (необязательное)

Напишите программу, которая сможет вычислять площади трех фигур (круг, треугольник и прямоугольник).
Тип фигуры запрашиваем через пользовательский ввод, после чего делаем запрос характеристик фигуры:

- если пользователь выбрал круг, запрашиваем его радиус,
- если треугольник – длины трех его сторон;
- если прямоугольник – длины двух его сторон.

Пример работы программы:  
1.  

```
Введите тип фигуры:
Круг

Введите радиус круга:
10
```
Результат:  
`Площадь круга: 314.16`  

2.  

```
Введите тип фигуры:
Треугольник

Введите длину стороны A:
2

Введите длину стороны B:
2

Введите длину стороны C:
3
```
Результат:  
`Площадь треугольника: 1.98`

````python
import math  # подключаем модуль для математических операций

# Запрашиваем тип фигуры
figure = input("Введите тип фигуры (круг, треугольник, прямоугольник): ").lower()

if figure == "круг":
    # Для круга нужен радиус
    radius = float(input("Введите радиус круга: "))
    area = math.pi * radius ** 2
    print(f"Площадь круга: {area:.2f}")

elif figure == "прямоугольник":
    # Для прямоугольника нужны две стороны
    a = float(input("Введите длину стороны A: "))
    b = float(input("Введите длину стороны B: "))
    area = a * b
    print(f"Площадь прямоугольника: {area:.2f}")

elif figure == "треугольник":
    # Для треугольника нужны три стороны
    a = float(input("Введите длину стороны A: "))
    b = float(input("Введите длину стороны B: "))
    c = float(input("Введите длину стороны C: "))
    
    # Проверяем, может ли существовать треугольник с такими сторонами
    if a + b > c and a + c > b and b + c > a:
        # Формула Герона
        p = (a + b + c) / 2  # полупериметр
        area = math.sqrt(p * (p - a) * (p - b) * (p - c))
        print(f"Площадь треугольника: {area:.2f}")
    else:
        print("Треугольник с такими сторонами не существует")

else:
    print("Неизвестный тип фигуры")
````
